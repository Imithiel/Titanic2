```python
import numpy as np #do roznych obliczen
import pandas as pd #do operacji na danych
import math

import matplotlib.pyplot as plt #do wykresow
import seaborn as sns #do jeszcze ladniejszych wykresow
%matplotlib inline 
pd.set_option('display.max_columns', 500)

r_filenameTSV = './TitanicCleaned.tsv'
print(r_filenameTSV)

# read the data
tsv_read = pd.read_csv(r_filenameTSV, sep='\t')

#pozbycie się duplikatów
db = tsv_read.drop_duplicates()
#db.head(60)
#db.sample(60)
#db.tail(60)
#db.shape
#db.to_csv('TitanicCleaned.tsv', sep="\t")

# Stopień przeżywalności

print(pd.crosstab(db["Survived"],db.Survived))

sns.set_palette("Set1")
sns.catplot(data=db,
           x='Survived',
           kind='count')
plt.show()

# Stopień przeżywalności względem klasy

print(pd.crosstab(db["Pclass"],db.Survived))

sns.catplot(data=db,
           x='Pclass',
           kind = 'count',
           hue= 'Survived')
plt.show()


# Wiek osoby, a wpływ na przeżycie

print(pd.crosstab(db["Age"]<18,db.Survived))

sns.distplot(db[db['Age'].notnull() & (db['Survived']==1)]['Age'], 
             kde_kws={"label": "Survived"}, 
             bins=10)
sns.distplot(db[db['Age'].notnull() & (db['Survived']==0)]['Age'], 
             kde_kws={"label": "Not Survived"}, 
             bins=10)
plt.show()


# Płeć, a przeżycie
print(pd.crosstab(db["Sex"],db.Survived))

sns.catplot(data=db,
           x='Sex',
           kind = 'count',
           hue= 'Survived')
plt.show()

# Embarked, a przeżycie
print(pd.crosstab(db["Embarked"],db.Survived))

sns.catplot(data=db,
           x='Embarked',
           kind = 'count',
           hue= 'Survived')
plt.show()

# Opłata, a przeżycie
print(pd.crosstab(db["Fare"]>50,db.Survived))

sns.catplot(data=db,
           x='Fare',
           kind = 'count',
           hue= 'Survived')
plt.show()

# SinSp/ Parch, a przeżycie
print(pd.crosstab(db["SibSp"],db.Survived))
print(pd.crosstab(db["Parch"],db.Survived))

sns.catplot(data=db,
           x='SibSp',
           kind = 'count',
           hue= 'Survived')
plt.show()

sns.catplot(data=db,
           x='Parch',
           kind = 'count',
           hue= 'Survived')
plt.show()


#summary(fileTitanic)


```
